```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('BBSI.JK.csv')

high_data = data['High']
low_data = data['Low']

total_high = high_data.sum()
total_low = low_data.sum()

percentages = [total_high, total_low]
labels = ['High', 'Low']
colors = ['red', 'blue']

# Membuat pie chart
plt.figure(figsize=(6, 6))
plt.pie(percentages, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart Persentase High dan Low')
plt.text(1,0.1,"Data High mendominasi secara signifikan")
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(6, 6))
plt.boxplot([high_data, low_data], labels=['High', 'Low'])
plt.title('Box Plot Data High dan Low')
plt.xlabel('Kolom Data')
plt.ylabel('Nilai')
plt.text(2.6,4000,"Variasi dalam data High lebih besar")
plt.show()
```
