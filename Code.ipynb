import pandas as pd import matplotlib.pyplot as plt

data = pd.read_csv(‘BAC.csv’)

open_data = data\[‘Open’\] close_data = data\[‘Close’\]

total_open, total_close = open_data.sum(), close_data.sum()

percentages = \[total_open, total_close\] labels, colors = \[‘Open’,
‘Close’\], \[‘lightblue’, ‘lightgreen’\]

date_data, profit, loss = data\[‘Date’\], open_data - close_data,
close_data - open_data

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 6))

# Box Plot

ax1.boxplot(\[open_data, close_data\], labels=labels) ax1.set(title=‘Box
Plot Data Open dan Close’, xlabel=‘Kolom Data’, ylabel=‘Nilai’)
ax1.text(0.5, 0.9, ‘Data cenderung miring ke kiri (negative skewness).’,
transform=ax1.transAxes, fontsize=10, color=‘blue’)

# Pie Chart

ax2.pie(percentages, labels=labels, colors=colors, autopct=‘%1.1f%%’,
startangle=140) ax2.set(title=‘Pie Chart Persentase Open dan Close’)
ax2.text(0.5, 0.9, ‘Perbandingan data relatif seimbang.’,
transform=ax2.transAxes, fontsize=10, color=‘green’)

# Area Chart

ax3.fill_between(date_data, open_data, close_data, where=(open_data \>
close_data), color=‘green’, alpha=0.5, label=‘Open’)
ax3.fill_between(date_data, open_data, close_data, where=(open_data \<=
close_data), color=‘red’, alpha=0.5, label=‘Close’) ax3.legend()
ax3.set(title=‘Area Chart Open vs Close’, xlabel=‘Tanggal’,
ylabel=‘Nilai’) ax3.text(0.5, 0.9, ‘Overlap Data: kedua data akan saling
berhimpit.’, transform=ax3.transAxes, fontsize=10, color=‘red’)

plt.tight_layout() plt.show()
