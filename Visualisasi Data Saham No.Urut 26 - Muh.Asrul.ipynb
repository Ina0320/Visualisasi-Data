### Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('tableau-colorblind10')
```

#### Load Data Sets PT Indofood Sukses Makmur Tbk (INDF.JK)

```python
dataasrul = pd.read_csv('INDF.JK.csv')
dataasrul.head()
```

##### Melakukan Extraksi Data Pada Colom Date Untuk Bulan nya Saja Agar Kita Bisa Membandingkan Berdasarkan Bulan

```python
dataasrul['Month'] = pd.to_datetime(dataasrul['Date']).dt.month_name()
dataasrul.head()
```

## Perbandingan (Comparison)

### Perbandingan Menggunakan Bar Chart

```python
# Melakukan GrouBy Pada Kolom Month Untuk Mencari/Membandingkan Total Volume Tiap-Tiap Bulan
Comparison = dataasrul.groupby('Month', as_index=False)['Volume'].sum()

plt.figure(figsize=(15, 4.5))
sns.barplot(x=Comparison['Month'],y=Comparison['Volume'])
plt.title('Comparison Volume By Month Pada Data Set PT Indofood Sukses Makmur Tbk (INDF.JK)', pad=12)

for i, asrul in enumerate(Comparison['Volume']):
    plt.text(i, asrul, str(asrul), ha='center', va='bottom')
plt.show()
```

###### Grafik/Visualisasi Di atas ini menunjukkan bahwa Volume Saham pada bulan May jauh lebih tinggi daripada bulan lainnya.

### Perbandingan Menggunakan Pie Chart

```python
plt.figure(figsize=(15, 4.5))
plt.pie(Comparison['Volume'], labels=Comparison['Month'], autopct='%1.1f%%')
plt.title('Comparison Volume By Month Pada Data Set PT Indofood Sukses Makmur Tbk (INDF.JK)', pad=12)
plt.show()
```

###### Pada Visualisasi Di atas menunjukan bahwa Bulan May mendominasi grafik dengan 12.0% dari total Volume Saham, sedangkan bulan lainya hanya menyumbang kurang dari 10.5%.
