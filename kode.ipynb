```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca data dari file CSV
data = pd.read_csv('BAC.csv')

# Mengambil kolom 'Open' dan 'Close'
open_data = data['Open']
close_data = data['Close']

# Menghitung total dari kedua kolom
total_open = open_data.sum()
total_close = close_data.sum()

# Persentase
percentages = [total_open, total_close]

# Label
labels = ['Open', 'Close']

# Warna
colors = ['lightblue', 'lightgreen']

# Membuat pie chart
plt.figure(figsize=(8, 8))
plt.pie(percentages, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart Persentase Open dan Close')
plt.text(1,1,"Data relatif seimbang")
plt.show()
```

```python
# Membuat box plot
plt.figure(figsize=(12, 8))
plt.boxplot([open_data, close_data], labels=['Open', 'Close'])
plt.title('Box Plot Data Open dan Close BAC')
plt.xlabel('Kolom Data')
plt.ylabel('Nilai')
plt.text(1.10,30.7,'Q2 (median) 30.7')
plt.text(1.10,28.7,'Q1 (kuartil bawah) 28.7')
plt.text(1.10,34,'Q3 (kuartil atas) 34')
plt.text(1.10,26.6,'Nilai Mininum 26.6')
plt.text(1.10,38,'Nilai Maksimum 38')
plt.show()
```

```python
# Mengambil kolom 'Date'
date_data = data['Date']

# Membuat area chart
plt.figure(figsize=(12, 6))
plt.fill_between(date_data, open_data, close_data, where=(open_data > close_data), color='green', alpha=0.5, label='Open')
plt.fill_between(date_data, open_data, close_data, where=(open_data <= close_data), color='red', alpha=0.5, label='Close')
plt.legend()
plt.title('Area Chart Open vs Close')
plt.xlabel('Tanggal')
plt.ylabel('Nilai')
plt.text(150,35,'kedua data saling berhimpit (overlap data)')
plt.show()
```
